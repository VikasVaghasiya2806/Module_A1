# Matrices

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Definitions and Notation
print("## 1. Definitions and Notation")

# Defining a matrix
matrix_a = np.array([[1, 2], [3, 4]])
print("Matrix A:\n", matrix_a)
```

```python
# 2. Types of Matrices
print("\n## 2. Types of Matrices")

# Square Matrix
square_matrix = np.array([[1, 2], [3, 4]])
print("Square Matrix:\n", square_matrix)

# Diagonal Matrix
diagonal_matrix = np.diag([1, 2, 3])
print("Diagonal Matrix:\n", diagonal_matrix)

# Identity Matrix
identity_matrix = np.eye(3)
print("Identity Matrix:\n", identity_matrix)

# Zero Matrix
zero_matrix = np.zeros((2, 3))
print("Zero Matrix:\n", zero_matrix)
```

```python
# 3. Matrix Operations
print("\n## 3. Matrix Operations")

# Matrix Addition
matrix_b = np.array([[5, 6], [7, 8]])
matrix_add = np.add(matrix_a, matrix_b)
print("Matrix Addition (A + B):\n", matrix_add)

# Matrix Subtraction
matrix_subtract = np.subtract(matrix_a, matrix_b)
print("Matrix Subtraction (A - B):\n", matrix_subtract)

# Matrix Multiplication
matrix_multiply = np.dot(matrix_a, matrix_b)
print("Matrix Multiplication (A * B):\n", matrix_multiply)
```

```python
# 4. Properties of Matrix Operations
print("\n## 4. Properties of Matrix Operations")

# Commutative Property (Addition)
print(np.add(matrix_a, matrix_b) == np.add(matrix_b, matrix_a))
commutative_addition = np.all(np.add(matrix_a, matrix_b) == np.add(matrix_b, matrix_a))
print("Commutative Property (Addition):", commutative_addition)

# Associative Property (Addition)
associative_addition = np.all(np.add(matrix_a, (matrix_b + matrix_a)) == np.add((matrix_a + matrix_b), matrix_a))
print("Associative Property (Addition):", associative_addition)

# Distributive Property
distributive_property = np.all(matrix_a.dot(matrix_b + matrix_a) == (matrix_a.dot(matrix_b) + matrix_a.dot(matrix_a)))
print("Distributive Property:", distributive_property)
```

```python
# 5. Transpose of a Matrix
print("\n## 5. Transpose of a Matrix")

matrix_transpose = np.transpose(matrix_a)
print("Transpose of Matrix A:\n", matrix_transpose)
```

```python
# 6. Inverse of a Matrix
print("\n## 6. Inverse of a Matrix")

# Calculating the Inverse
matrix_inverse = np.linalg.inv(matrix_a)
print("Inverse of Matrix A:\n", matrix_inverse)

# Verifying the Inverse
identity_check = np.dot(matrix_a, matrix_inverse)
print("A * A^-1 (Identity Matrix):\n", identity_check)
```

```python
# 7. Determinants and their Properties
print("\n## 7. Determinants and their Properties")

# Determinant of a Matrix
determinant_a = np.linalg.det(matrix_a)
print("Determinant of Matrix A:", determinant_a)

# Properties of Determinants
# Determinant of the product of matrices
product_determinant = np.linalg.det(np.dot(matrix_a, matrix_b))
property_check = determinant_a * np.linalg.det(matrix_b)
print("Det(A*B) =", product_determinant)
print("Det(A) * Det(B) =", property_check)
```

# Vector Spaces

```python
b1 = np.array([5,0])
b2 = np.array([1,1])
origin = np.zeros(2)
fig = plt.figure()
plt.quiver(*origin,*b1,color='r')
plt.quiver(*origin,*b2,color='b')
```

```python
# Function to plot vectors
def plot_vectors(vectors, colors, labels, title):
    origin = np.zeros((3, len(vectors)))
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    for i in range(len(vectors)):
        ax.quiver(*origin[:, i], *vectors[i], color=colors[i], label=labels[i])
    ax.set_xlim([-3, 3])
    ax.set_ylim([-3, 3])
    ax.set_zlim([-3, 3])
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    plt.legend()
    plt.title(title)
    plt.show()
```

```python
# 1. Vector operations
v1 = np.array([1, 0, 1])
v2 = np.array([1, 1, 0])
v_add = v1 + v2
v_sub = v1 - v2
v_dot = np.dot(v1, v2)
v_cross = np.cross(v1, v2)

print("Vector Addition: ", v_add)
print("Vector Subtraction: ", v_sub)
print("Dot Product: ", v_dot)
print("Cross Product: ", v_cross)

plot_vectors([v1, v2, v_cross], ['r', 'b', 'g'], ['v1', 'v2', 'v1 cros v2'], "Vector Addition")

```

```python
plot_vectors([v1, v2, v_dot], ['r', 'b', 'g'], ['v1', 'v2', 'v1dotv2'], "Vector dot")
```

```python
# 2. Basis
basis = [np.array([1, 0, 0]), np.array([0, 1, 0]), np.array([0, 0, 1])]
plot_vectors(basis, ['r', 'g', 'b'], ['i', 'j', 'k'], "Standard Basis")
```

```python
# 3. Change of basis
A = np.array([[1, 1, 0], [0, 1, 1], [1, 0, 1]])
new_basis = np.dot(A, basis)
plot_vectors(new_basis, ['r', 'g', 'b'], ['b1', 'b2', 'b3'], "New Basis after Change")
```

```python
# 4. Subspace and Span

# Define the vectors v1 and v2
v1 = np.array([1, 0, 0])
v2 = np.array([0, 0, 1])

# Generate a set of vectors in the span of v1 and v2
# Here we take linear combinations of v1 and v2
c1_values = np.linspace(-1, 1, 5)
c2_values = np.linspace(-1, 1, 5)
span_vectors = []

for c1 in c1_values:
    for c2 in c2_values:
        span_vectors.append(c1 * v1 + c2 * v2)

# Convert list to array for easy plotting
span_vectors = np.array(span_vectors)

# Plot the span of v1 and v2
plot_vectors([v1, v2], ['r', 'g'], ['v1', 'v2'], "Basis Vectors v1 and v2")
plot_vectors(span_vectors, ['b'] * len(span_vectors), ['span(v1,v2)'] * len(span_vectors), "Subspace Spanned by v1 and v2")
```

```python
# 5. Orthogonality check
u = np.array([1, 0, 0])
v = np.array([0, 1, 0])
orthogonality = np.dot(u, v) == 0
print("Are vectors u and v orthogonal?", orthogonality)

plot_vectors([u, v], ['r', 'g'], ['u', 'v'], "Orthogonal Vectors")
```

```python
# 6. Linear independence check
# If the determinant of the matrix formed by the vectors is non-zero, they are linearly independent.
v1 =  np.array([1, 0, 0])
v2 =  np.array([1, 1, 0])
v3 =  np.array([2, 1, 0])
matrix = np.array([v1, v2, v3])
lin_independence = np.linalg.det(matrix) != 0
print("Are the vectors linearly independent?", lin_independence)
plot_vectors([v1, v2, v3], ['r', 'g','b'], ['v1', 'v2','v3'], "Basis Vectors v1 and v2")
```
