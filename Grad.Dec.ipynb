# Least Square Example

```python
xp = np.random.rand(10)#np.linspace(0,10,10)
yp = np.random.rand(10)#2*xp+1+0.1*np.random.randn(10)
zp = 2*xp+3*yp+1
```

```python
A = np.array([xp,yp,np.ones(10)]).T
```

```python
A.shape
```

```python
b = np.reshape(zp,[10,1])
b
```

```python
x = np.linalg.inv(A.T @ A) @ A.T @ b
```

```python
a = x[0]
bb = x[1]
c = x[2]
a,b,c
```

```python
x = np.array([[3],[3],[3]])
len = 0.001
cost = np.linalg.norm(A @ x - b,2)
for i in range(1000):
    cost = np.linalg.norm(A @ x - b,2)
    grad = A.T @ A @ x - A.T @ b
    x = x - len*grad
    if i%100==0:
        print(f'cost = {cost} a = {x[0]} c = {x[1]} d = {x[2]}')
```
