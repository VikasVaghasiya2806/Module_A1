```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_vectors2d(vectors, colors,labels):
    origin = np.zeros((2, len(vectors)))
    fig, ax = plt.subplots()
    a = 0
    for i in range(len(vectors)):
        ax.quiver(*origin[:, i], *vectors[i], color=colors[i],scale=1, scale_units='xy', angles='xy',label=labels[i])
        if a<np.max(vectors[i]):
            a = np.max(vectors[i])
    a = a+1
    ax.set_xlim([-a, a])
    ax.set_ylim([-a, a])
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.legend()
    plt.show()
```

```python
# Function to plot vectors
def plot_vectors(vectors, colors,labels):
    origin = np.zeros((3, len(vectors)))
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    a = 0
    for i in range(len(vectors)):
        ax.quiver(*origin[:, i], *vectors[i], color=colors[i],label=labels[i])
        if a<np.max(vectors[i]):
            a = np.max(vectors[i])
    a = a+1
    ax.set_xlim([-a, a])
    ax.set_ylim([-a, a])
    ax.set_zlim([-a, a])
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    #ax.legend()
    plt.show()
```
