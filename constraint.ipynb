```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[0.7,0.2],[.1,.6]])
x = np.array([[.4],[.6]])
b_n = np.array([[.5],[.6]]) 
b = np.array([[.4],[.4]]) 
```

```python
x_e = np.linalg.inv(A.T @ A) @ A.T @ b_n
```

```python
m,n = A.shape
A_c = np.zeros([m+1,n])
A_c[:-1,:] = A
A_c[-1:,:] = np.ones([1,n])
b_nc = np.zeros([m+1,1])
b_nc[:-1,:] = b_n
b_nc[-1:,:] = 1
```

```python
x_ec = np.linalg.inv(A_c.T @ A_c) @ A_c.T @ b_nc
```

```python
x_e
```

```python
fig, ax = plt.subplots(figsize=(5, 5)) 
ax.axis('equal')
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
z = np.zeros(2)
b_ne = A @ x_e
b_nc = A_c @ x_ec
b_nc = b_nc[:-1]
ax.quiver(*z,*A[:,0],color='k',scale=1, scale_units='xy', angles='xy',label = 'A[:,0]')
ax.quiver(*z,*A[:,1],color='k',scale=1, scale_units='xy', angles='xy',label = 'A[:,1]')
ax.quiver(*z,*b,color='b',scale=1, scale_units='xy', angles='xy',label = 'original')
ax.quiver(*z,*b_n,color='r',scale=1, scale_units='xy', angles='xy',label ='noisy')
ax.quiver(*z,*b_ne,color='y',scale=1, scale_units='xy', angles='xy',label ='least_square')
ax.quiver(*z,*b_nc,color='g',scale=1, scale_units='xy', angles='xy',label ='constraint')
plt.legend()
plt.show()
```
