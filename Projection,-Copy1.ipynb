# Projection

```python
import numpy as np
import matplotlib.pyplot as plt 
import time
```

## Point to line

```python
P = np.array([3,6])#point on y = 2x
x = np.linspace(-5,5,10)
y = x+1
plt.figure()
plt.scatter(P[0],P[1],marker='*',label='P') 
plt.plot(x,y,color='r',label='line:x=y')
plt.legend()
plt.show()
```

```python
def point_proj(P,A,B,C):#P =point , Ax+By+C=0 line
    #direction of line
    a = np.array([-C/A,0])-np.array([0,-C/B])
    b = P-np.array([0,-C/B]) 
    return np.array([0,-C/B])+((np.dot(a,b)/np.dot(a,a))*a)
```

```python
P2 = point_proj(P,1,-1,1)
```

```python
fig, ax = plt.subplots(figsize=(5, 5)) 
ax.axis('equal')
ax.scatter(P[0],P[1],marker='*',label='P') 
ax.plot(x,y,color='r',label='x-y-1=0')
ax.scatter(P2[0],P2[1],marker='o',label='P2')
ax.quiver(*P2,*P)
plt.legend()
plt.show()
```

```python
P = np.array([0,3])#point on y = 2x+1
x = np.linspace(-5,5,10)
y = x+1
y2 = 2*x+3
#plt.figure()
fig, ax = plt.subplots(figsize=(5, 5)) 
ax.axis('equal')
ax.scatter(P[0],P[1],marker='*',label='P') 
ax.plot(x,y,color='r',label='line:y=x+1')
ax.plot(x,y2,color='g',label='line:y=2x+1')
plt.legend()
plt.show()
```

```python
#point on y=2x+1 > projection on y=x+1
P2 = point_proj(P,1,-1,1)
fig, ax = plt.subplots(figsize=(5, 5)) 
ax.axis('equal')
ax.set_xticks(np.linspace(-5, 5, 10), minor=True)
ax.set_yticks(np.linspace(-5, 5, 10), minor=True)
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.tick_params(axis='x', colors='red')
ax.tick_params(axis='y', colors='blue')
ax.scatter(P[0],P[1],marker='*',label='P') 
ax.plot(x,y,color='r',label='line:y=x+1')
ax.plot(x,y2,color='g',label='line:y=2x+1')
ax.scatter(P2[0],P2[1],marker='o',label='P2') 
for i in range(50):
    P = point_proj(P2,2,3,-1)
    ax.scatter(P[0],P[1],marker='*')
    if np.all(P2==P):
        break
    P2 = point_proj(P,1,-1,1)
    ax.scatter(P2[0],P2[1],marker='o')
    
plt.legend()
plt.show()
```
